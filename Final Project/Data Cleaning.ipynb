```python
#Extract category info from json file 
def extract_category_name(path):
  with open(path) as json_file:
    data = json.load(json_file)
    category_list = []
    for row in data['items']:
      id_number = row['id']
      category_name = row['snippet']['title'] 
      category_info = (id_number, category_name)
      category_list.append(category_info)
    return(dict(category_list))
```

```python
#Fetch categories info for three countries 
cac = extract_category_name("CA_category_id.json")
usc = extract_category_name("US_category_id.json")
gbc = extract_category_name("GB_category_id.json")
```

```python
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import RegexpTokenizer
import re
```

```python
## url
def delurl(string):
    result = re.sub(r"http\S+", "", string)
    output = ''.join(result)
    return output
```

```python
## stopwords
stop_words = set(stopwords.words('english')) 
def delstopwords(string): 
    word_tokens = word_tokenize(string) 
    result = [w for w in word_tokens if not w in stop_words]
    result = []  
  
    for w in word_tokens:  
        if w not in stop_words:  
            result.append(w)
    output = ' '.join(result)
    return output
```

```python
## punctuations
def delpunct(string):
    tokenizer = RegexpTokenizer(r'\w+')
    res = tokenizer.tokenize(string)
    output = ' '.join(res)
    return output
```

```python
def delete(string):
    res = delurl(string)
    res = delpunct(res)
    res = delstopwords(res)
    return res
    
```
